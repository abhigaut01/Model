## RFM Segmentation

```python
#Method-1: Creating quartiles

#data['R_Quartile'] = pd.qcut(data['recency'], 4, labels=False)
#data['F_Quartile'] = pd.qcut(data['frequency'], 4, labels=False)
#data['M_Quartile'] = pd.qcut(data['monitory_value'], 4, labels=False)
```

```python
#Method-2: Creating quartiles

quantiles = data[['frequency','monitory_value','recency' ]].quantile(q=[0.25,0.5,0.75])
quantiles
```

```python
quantiles = quantiles.to_dict()
quantiles
```

```python
#We create two classes for the RFM segmentation since, being high recency is bad, while high frequency and monetary value is good.
# Arguments (x = value, p = recency, monetary_value, frequency, d = quartiles dict)
# For recency
def RClass(x,p,d) :
    if x<=d[p][0.25]:
        return 1
    elif x<=d[p][0.50]:
        return 2
    elif x<=d[p][0.75]:
        return 3
    else:
        return 4

# For frequency and monitory
def FMClass(x,p,d) :
    if x<=d[p][0.25]:
        return 4
    elif x<=d[p][0.50]:
        return 3
    elif x<=d[p][0.75]:
        return 2
    else:
        return 1
```

```python
data['R_Quartile'] = data['recency'].apply(RClass, args=('recency',quantiles,))
data['F_Quartile'] = data['frequency'].apply(FMClass, args=('frequency',quantiles,))
data['M_Quartile'] = data['monitory_value'].apply(FMClass, args=('monitory_value',quantiles,))
```

```python
data.head()
```

```python
data1=data.assign(R_seg=np.nan, FM_seg =np.nan)
data1.R_seg[data['R_Quartile']==4] = "3.Churn"
data1.R_seg[data['R_Quartile']==2] = "2.At Risk"
data1.R_seg[data['R_Quartile']==3] = "2.At Risk"
data1.R_seg[data['R_Quartile']==1] = "1.Active"

data1.FM_seg[(data['F_Quartile']==3) & (data['M_Quartile']==1)] = "2.Gold"
data1.FM_seg[(data['F_Quartile']==4) & (data['M_Quartile']==1)] = "2.Gold"
data1.FM_seg[(data['F_Quartile']==2) & (data['M_Quartile']==2)] = "2.Gold"
data1.FM_seg[(data['F_Quartile']==3) & (data['M_Quartile']==2)] = "2.Gold"
data1.FM_seg[(data['F_Quartile']==4) & (data['M_Quartile']==4)] = "1.Silver"
data1.FM_seg[(data['F_Quartile']==1) & (data['M_Quartile']==3)] = "2.Gold"
data1.FM_seg[(data['F_Quartile']==2) & (data['M_Quartile']==3)] = "2.Gold"
data1.FM_seg[(data['F_Quartile']==3) & (data['M_Quartile']==3)] = "1.Silver"
data1.FM_seg[(data['F_Quartile']==4) & (data['M_Quartile']==3)] = "1.Silver"
data1.FM_seg[(data['F_Quartile']==1) & (data['M_Quartile']==4)] = "2.Gold"
data1.FM_seg[(data['F_Quartile']==2) & (data['M_Quartile']==4)] = "1.Silver"
data1.FM_seg[(data['F_Quartile']==3) & (data['M_Quartile']==4)] = "1.Silver"
data1.FM_seg[(data['F_Quartile']==4) & (data['M_Quartile']==4)] = "0.Standard"

data1.FM_seg.fillna("4.Premium", inplace=True)
```
